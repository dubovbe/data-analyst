# Пропуски в данных высота потолков (ceiling_height)

Параметр "высота потолков" у большенства квартир стандратный. В данных этот параметр вероятно вводился продавцом. Предпологаем, что, если данные опущены, значит этот параметр у квартиры стандартный.

Найдем медиальную величину параметра:

```python
median_ceiling_height = df['ceiling_height'].median()
median_ceiling_height
```

Квартир с такой высотой потолов всего 4.66%:

```python
df.query('ceiling_height == @median_ceiling_height')['ceiling_height'].count()/df['ceiling_height'].count()*100
```

```python
mode_ceiling_height = df['ceiling_height'].mode()
mode_ceiling_height
```

Процент квартир с самой популярной высотой потолков 24.23%:

```python
df.query('ceiling_height == 2.50')['ceiling_height'].count()/df['ceiling_height'].count()*100
```

```python
df['ceiling_height'].fillna(2.50, inplace=True)
```

Можно сделать предположение, что высота потолка 2.5 м является стандартной. Возможно продавец просто не посчитал нужным отражать этот параметр у себя в объявлении. Стоит заменить пропуски в 'ceiling_height' именно самым часто встречающимся значением, а не медианой.

# Пропуски в данных "жилая площадь в квадратных метрах" (living_area):

```python
df.query('living_area.isna()')
```

Такие пропуски можно заполнить на основании значения медианы или среднего арифметического от площади жилой зоны, отноносительно числа комнат в квартире. 

```python
r0 = (df['rooms'] == 0)
r1 = (df['rooms'] == 1)
r2 = (df['rooms'] == 2)
r3 = (df['rooms'] == 3)
r4 = (df['rooms'] == 4)
r5 = (df['rooms'] == 5)
r6 = (df['rooms'] == 6)
r7 = (df['rooms'] == 7)
r9 = (df['rooms'] == 9)
df.loc[r0,'living_area'] = df.loc[r0,'living_area'].fillna(df.loc[r0,'living_area'].median())
df.loc[r1,'living_area'] = df.loc[r1,'living_area'].fillna(df.loc[r1,'living_area'].median())
df.loc[r2,'living_area'] = df.loc[r2,'living_area'].fillna(df.loc[r2,'living_area'].median())
df.loc[r3,'living_area'] = df.loc[r3,'living_area'].fillna(df.loc[r3,'living_area'].median())
df.loc[r4,'living_area'] = df.loc[r4,'living_area'].fillna(df.loc[r4,'living_area'].median())
df.loc[r5,'living_area'] = df.loc[r5,'living_area'].fillna(df.loc[r5,'living_area'].median())
df.loc[r6,'living_area'] = df.loc[r6,'living_area'].fillna(df.loc[r6,'living_area'].median())
df.loc[r7,'living_area'] = df.loc[r7,'living_area'].fillna(df.loc[r7,'living_area'].median())
df.loc[r9,'living_area'] = df.loc[r9,'living_area'].fillna(df.loc[r9,'living_area'].median())
```

# Пропуски в данных "площадь кухни в квадратных метрах" (kitchen_area):

Пропуски в данных 'kitchen_area' заполним через ср. коэф. отношения 'kitchen_area'/'living_area'

```python
ka_mean = (df['kitchen_area']/df['total_area']).mean()
ka_mean
```

```python
df['kitchen_area'].fillna(df['total_area']*ka_mean, inplace=True)
```

# Пропуски в данных об расстояниях до ближайших объектов (airports_nearest, cityCenters_nearest, parks_nearest, ponds_nearest):

Восстановить такие пропуски можно, опираясь на данные сводной таблицы по районам. Оставшиеся пропуски можно будет оставить.

```python
grp = df.groupby(['locality_name'])
df.loc[df['locality_name'] != 'Санкт-Петербург', 'airports_nearest'] = grp.airports_nearest.apply(lambda x: x.fillna(x.median()))
df.loc[df['locality_name'] != 'Санкт-Петербург', 'cityCenters_nearest'] = grp.cityCenters_nearest.apply(lambda x: x.fillna(x.median()))
```

# Остальные пропуски 

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
df['locality_name'].unique()
```

```python
len(df['locality_name'].unique().tolist())
```

```python
location_status = [
    'садоводческое некоммерческое товарищество',
    'садовое товарищество',
    'поселок при железнодорожной станции',
    'поселок городского типа',
    'коттеджный поселок',
    'городской поселок',
    'поселок станции',
    'деревня',
    'поселок',
    'село',
]

def locality_name_corr(row):
    locality_name = row['locality_name']
    if type(locality_name) == str:
        locality_name = locality_name.replace("ё", "е")
        for status in location_status:
            if status in locality_name:
                locality_name = locality_name[len(status):].strip()  
    return locality_name
```

```python
df['locality_name'] = df.apply(locality_name_corr, axis=1)
```

```python
len(df['locality_name'].unique().tolist())
```

### Этаж квартиры, относительно высоты дома

```python
def floor_type(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == floors_total:
        return 'последний'
    if floor == 1:
        return 'первый'
    return 'другой'
```

```python
df['floor_type'] = df.apply(floor_type, axis=1)
```
